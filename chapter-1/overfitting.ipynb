```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeRegressor
import seaborn as sns
sns.set_theme(style="white", palette='bright')
```

```python
x = np.arange(0, 10.1, 0.5)
forecast_x = np.arange(0, 12.1, 0.5)
```

```python
signal = 0.3*x + 5
noise = np.random.uniform(-1, 1, len(x))
y = signal + noise
```

```python
tree = DecisionTreeRegressor(max_depth=10)
tree = tree.fit(x.reshape(-1, 1), y.reshape(-1, 1))
```

```python
y_pred = tree.predict(forecast_x.reshape(-1, 1))
```

```python
tree.predict(np.array([12]).reshape(1, -1))
```

```python
sns.regplot(x=x, y=y, ci=None, line_kws={'color': 'g', 'label': 'signal'})
plt.plot(forecast_x, y_pred, linestyle='--')
plt.legend(labels=['data', 'true signal', 'model fit'])
plt.grid()
plt.savefig("figures/ch1_overfitting_example.png", dpi=600)
```
