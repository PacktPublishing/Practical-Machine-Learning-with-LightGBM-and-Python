```python
import numpy as np
from matplotlib import pyplot as plt

import seaborn as sns
sns.set_theme(style="white", palette='bright')
```

```python
f = lambda x: np.sin(x) * x * x + np.cos(x) * np.cos(x)
```

```python
X = np.arange(1.9, 6.6, 0.01)
```

```python
Y = f(X)
```

```python
GD_X = np.array([3, 3.5, 4.0, 4.5, 5])
GD_Y = f(GD_X)
GD_labels = ['x_0', 'x_1', 'x_2', 'x_3', 'x_4']
```

```python
sns.lineplot(x=X, y=Y)
plt.plot(GD_X, GD_Y, 'o:', label='gradient descent steps')
plt.legend()
plt.ylim([-28, 10])

for i, label in enumerate(GD_labels):
    plt.annotate(f'${label}$', (GD_X[i] - 0.1, GD_Y[i] - 2.5))
plt.savefig("figures/ch2_gradient_descent_example.png", dpi=600)
```
